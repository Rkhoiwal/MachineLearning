```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('/Users/mack/dev/ml/notebook/wine.csv')
data.head()
```

```python
data.describe()
```

```python
data.columns = ["Class","Alcohol", "Malic acid", "Ash" ,"Alcalinity of ash", "Magnesium","Total phenols","Flavanoids","Nonflavanoid phenols"
,"Proanthocyanins","Color intensity","Hue","OD280/OD315 of diluted wines","Proline"]
```

```python
from sklearn.utils import shuffle
data = shuffle(data, random_state=15).reset_index(drop=True)
```

```python
from sklearn.model_selection import train_test_split
y = data['Class']


from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()
y_enc = enc.fit_transform(y)
from keras.utils import to_categorical
y_cat = to_categorical(y_enc)


X = data.drop('Class', axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y_cat, test_size=0.25)
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
```

```python
model.fit(X_train, y_train)
```

```python
predicted = model.predict(X_test)
```

```python
print (np.mean(predicted == y_test))
```
